```python
#!/usr/bin/env python3

from skimage import io
import matplotlib.pyplot as plt
import numpy as np
from skimage.color import rgb2gray
from skimage.filters import threshold_local, threshold_otsu

img = io.imread('/home/ipsita/Downloads/image/8068.jpg')
img_grey = rgb2gray(img)
plt.imshow(img_grey, cmap=plt.cm.gray, interpolation='nearest')
plt.show()
```

```python
#global threshold
global_thresh = threshold_otsu(img_grey)
binary_global = img_grey > global_thresh

#local threshold
block_size = 23
local_thresh = threshold_local(img_grey, block_size)
binary_local = img_grey > local_thresh
```

```python
plt.hist(img_grey.ravel(), bins=256)
plt.title('Global thresh histogram')
plt.axvline(global_thresh, color ='r')
plt.show()
```

```python
plt.imshow(binary_global, cmap=plt.cm.gray)
plt.title('Global thresholding')
plt.show()
```

```python
plt.imshow(binary_local, cmap=plt.cm.gray)
plt.title('Local thresholding')
plt.show()
```
