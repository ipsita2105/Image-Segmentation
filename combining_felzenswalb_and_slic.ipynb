```python
#!/usr/bin/env python3

from skimage import data, io, segmentation, color
from skimage.future import graph
from matplotlib import pyplot as plt
from skimage.color import rgb2gray
from skimage.feature import canny
from skimage.morphology import watershed, disk
from skimage.morphology import dilation, disk, erosion, opening, closing
import numpy as np
from skimage.measure import label
from skimage.segmentation import felzenszwalb

img = io.imread('/home/ipsita/Downloads/image/87015.jpg')
img_grey = rgb2gray(img)
plt.imshow(img)
plt.show()
```

```python
img_grey = rgb2gray(img)

seg1 = segmentation.slic(img, compactness=30, n_segments=200)
seg2 = segmentation.felzenszwalb(img, scale=500, sigma=0.5, min_size=50)


out_t2 = color.label2rgb(seg2, img, kind='avg')
out_t1 = color.label2rgb(seg1, img, kind='avg')
```

```python
plt.title('SLIC')
plt.imshow(out_t1)
plt.show()
```

```python
plt.title('Felzenswalb')
plt.imshow(out_t2)
plt.show()
```

```python
seg_mix = segmentation.join_segmentations(seg1, seg2)

out_mix = color.label2rgb(seg_mix, img, kind='avg')

plt.title('Intersection')
plt.imshow(out_mix)

plt.show()
```
