```python
from skimage import io
import matplotlib.pyplot as plt
import numpy as np
from skimage.color import rgb2gray
from skimage.feature import canny
from scipy import ndimage as ndi
from skimage import morphology
from skimage.morphology import dilation, disk, erosion, opening, closing

img = io.imread('/home/ipsita/Downloads/image/8068.jpg')
img_grey = rgb2gray(img)
plt.title('Original')
plt.imshow(img_grey, cmap=plt.cm.gray, interpolation='nearest')
plt.show()
```

```python
#delinate the contour using edge-based segmentation
#canny edge-detector
edges = canny(img_grey)

plt.title('Canny detector')
plt.imshow(edges, cmap=plt.cm.gray, interpolation='nearest')
plt.show()
```

```python
#fill countours using mathematical morphology
fill_img = ndi.binary_fill_holes(edges)

plt.title('filling the holes')
plt.imshow(fill_img, cmap=plt.cm.gray, interpolation='nearest')
plt.show()
```

```python
#remove small objects by setting a minimum for valid objects
clean_img = morphology.remove_small_objects(fill_img, 21)

plt.title('remove small objects')
plt.imshow(clean_img, cmap=plt.cm.gray, interpolation='nearest')

plt.show()
```

```python
#Now applying a morphological transformation

selem = disk(6)
dialated = opening(img_grey, selem)

#dilated image
plt.imshow(dialated)
plt.show()
```

```python
#dilated edges
edges = canny(dialated)
plt.imshow(edges, cmap=plt.cm.gray, interpolation='nearest')
plt.show()
```

```python
#dilated holes
fill_img = ndi.binary_fill_holes(edges)
plt.imshow(fill_img, cmap=plt.cm.gray, interpolation='nearest')
plt.show()
```
